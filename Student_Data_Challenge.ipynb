```python
#Add in imports

import pandas as pd
import os
import numpy as np
```

```python
#join CSV file
student_data = os.path.join('new_full_student_data.csv')
```

```python
student_df = pd.read_csv(student_data)
```

```python
#Check to make sure data is correct

student_df.head()
```

```python
#ID how many null values we have

student_df.isnull().sum()
```

```python
student_df = student_df.dropna()
```

```python
#ID Duplicates

student_df.duplicated()
```

```python
student_df.duplicated().sum()
```

```python
#Remove Duplicates 

student_df = student_df.drop_duplicates()
student_df
```

```python
student_df.isnull().sum()
```

```python
#Look at the types of data

student_df.dtypes
```

```python
# Remove "th after grade number

student_df.loc[:, "grade"] = student_df.loc[:, "grade"].str.replace("th", "")
```

```python
student_df
```

```python
#Turn grade into an integer

student_df.loc[:, "grade"] = student_df.loc[:, "grade"].astype(int)
```

```python
student_df.dtypes
```

```python
#Show the stats of the data

student_df.describe()
```

```python
#Find the mean 

student_df.mean()
```

```python
#Store the min reading score
min_reading_score = student_df["reading_score"].min()
```

```python
#Display the grade column using loc

student_df.loc[:, "grade"]
```

```python
#Display the first three rows of columns 3,4, and 5. 

student_df.iloc[[0, 1, 2], 3 : 6]
```

```python
#Select rows for grade 9 and display summary statistics

summary_grade_nine_df = student_df.loc[student_df["grade"] == 9]
summary_grade_nine_df.describe()                                       
```

```python
#Store the row with min overall reading score by using loc and min_reading_score from variable 3. 

min_reading_score = student_df["reading_score"].min()

min_reading_row = student_df.loc[student_df["reading_score"] == min_reading_score]

min_reading_row
```

```python
#Select all the reading scores from 10th grade at Dixon High School

tenth_grade_reading_dixon = student_df.loc[student_df["grade"] == 10]
tenth_grade_reading_dixon = tenth_grade_reading_dixon[student_df["school_name"] == "Dixon High School"]
tenth_grade_reading_dixon
```

```python
#Find the mean reading score for all the students in Grades 11 and 12 combined
combined_reading_score = student_df.loc[(student_df["grade"] == 11) | (student_df["grade"] == 12)][["reading_score"]].mean()
combined_reading_score.mean()
```

```python
#Display the average budget for each school type by using the groupby and mean functions
student_df.groupby(["school_type"])[["school_budget"]].mean()
```

```python
#Find the total number of students at each school, sort numbers largest to smallest using the groupby, count, and sort_values

student_df.groupby(["school_name"])[["student_name"]].count().sort_values(by='student_name', ascending = False)
```

```python
#Find the average math score by grade for each school type by using the groupby and mean functions
student_df.groupby(["grade", "school_type"])[["math_score"]].mean()
```

```python
#Added - Avg reading_score by school type and grade. 
student_df.groupby(["grade", "school_type"])[["reading_score"]].mean()
```
